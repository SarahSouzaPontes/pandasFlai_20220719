```python
#Teste1 comprador idade acima da media de idade
def aux(x):
  if x > dataset["Age"].mean():
    return 1
  else:
    return 0

dataset ['Teste1'] = dataset ['Age'].apply(aux)
```

```python
 dataset["Age"].mean()
```

```python
dataset
```

```python
dataset['Purchased'] == dataset['Teste1']
```

```python
(dataset['Purchased'] == dataset['Teste1']).mean()
```

```python
#Teste2 vou inferir como comprador aqueles que tem salario acima da media
dataset["Teste2"]= 0
dataset.loc[dataset ["EstimatedSalary"] > dataset ["EstimatedSalary"].mean(), "Teste2"] = 1
```

```python
dataset.head()
```

```python
#acuracia
(dataset['Purchased']== dataset ['Teste2']).mean()
```
